```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
netflix_file = "../ETL/Netflix Shows.csv"
netflix_df = pd.read_csv(netflix_file, encoding='latin-1')
netflix_df.head()
```

```python
#Transform  data 1

netflix_df = netflix_df[['title','release year','rating']]


netflix_df.set_index("title", inplace=True)



netflix_df.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/netflix_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
netflix_df.to_sql(name='netflix', con=engine, if_exists='append', index=True)
```

```python
pg_table = pd.read_sql_query('select * from netflix', con=engine)
pg_table
```

```python
# IF you have additional data to add, follow this code

netflix_file2 = "../ETL/netflix_titles.csv"
netflix_df2 = pd.read_csv(netflix_file2)
netflix_df2.head()
netflix_df2
```

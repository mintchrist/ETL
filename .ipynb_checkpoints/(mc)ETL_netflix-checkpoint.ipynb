```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
netflix_file = "../ETL/Netflix Shows.csv"
netflix_df = pd.read_csv(netflix_file, encoding='latin-1')
netflix_df.head()
```

```python
netflix_file2 = "../ETL/netflix_titles.csv"
netflix_df2 = pd.read_csv(netflix_file2)
netflix_df2.head()
```

```python
#Transform  data 1

netflix_df = netflix_df[['title','release year','rating']]


netflix_df.set_index("title", inplace=True)



netflix_df.head()
```

```python
#Transform data 2

netflix_df2 = netflix_df2[['title','release_year','rating']]
netflix_df2 = netflix_df2.rename(columns = {'release_year': 'release year'})

netflix_df2.set_index("title", inplace=True)


netflix_df2.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/netflix_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
# load data top confirm

netflix_df.to_sql(name='netflix', con=engine, if_exists='append', index=True)
```

```python
netflix_df2.to_sql(name='netflix', con=engine, if_exists='append', index=True)
```

```python

pd.read_sql_query('select * from netflix', con=engine)
```

```python
netflix_df3 = pd.concat([netflix_df,netflix_df2])
```

```python
netflix_df3.reset_index().drop_duplicates(subset='title')
```
